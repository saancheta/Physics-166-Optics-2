### Anamorphic property of FT of different 2D patterns

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])
```

### Tall rectangle aperture

```python
tall = mpimg.imread('tall.bmp')
gray= rgb2gray(tall)
Tall = gray
plt.imshow(Tall, cmap = 'gray')
```

```python
F_Tall = abs(np.fft.fft2(Tall))
F_Tall1 = abs(np.fft.fftshift(F_Tall))
plt.imshow(F_Tall1, cmap = 'gray')
```
