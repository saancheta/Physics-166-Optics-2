### Rotation property of the FT

```python
import matplotlib.pyplot as plt
import numpy as np
```

### At frequency f = 5:

```python
f = 5

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
Sin = np.sin(2*np.pi*f*X)
#z = sin(2*%pi*f*X);

plt.imshow(Sin, cmap = 'gray')

```

```python
F_Sin = abs(np.fft.fft2(Sin))
F_Sin1 = np.fft.fftshift(F_Sin)
plt.imshow(F_Sin1, cmap = 'gray')
```

### Increasing the frequency:

```python
f = 100

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
Sine = np.sin(2*np.pi*f*X)

plt.imshow(Sine, cmap = 'gray')
```

```python
F_Sine = abs(np.fft.fft2(Sine))
F_Sine1 = np.fft.fftshift(F_Sine)
plt.imshow(F_Sine1, cmap = 'gray')
```

### Rotating the sinusoid:

```python
f = 5
theta = 30

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
RSin = np.sin(2*np.pi*f*(Y*np.sin(theta) + X*np.cos(theta)))

plt.imshow(RSin, cmap = 'gray')
```

```python
F_RSin = abs(np.fft.fft2(RSin))
F_RSin1 = np.fft.fftshift(F_RSin)
plt.imshow(F_RSin1, cmap = 'gray')
```

### Combination of sinusoids in X and Y:

```python
f = 5
theta = 30

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
CombSin = (np.sin(2*np.pi*(f*X)))*(np.sin(2*np.pi*(f*Y)))

plt.imshow(CombSin, cmap = 'gray')
```

```python
F_CombSin = abs(np.fft.fft2(CombSin))
F_CombSin1 = np.fft.fftshift(F_CombSin)
plt.imshow(F_CombSin1, cmap = 'gray')
```

### Rotating this combination of sinusoids in X and Y:

```python
f = 5
theta = 20

nx= 100
ny= 100
x = np.linspace(-1,1,nx)
y = np.linspace(-1,1,ny)
X,Y = np.meshgrid(x,y)
RCombSin = (np.sin(2*np.pi*(f*(X*np.cos(theta)))))*(np.sin(2*np.pi*(f*(Y*np.cos(theta)))))

plt.imshow(RCombSin, cmap = 'gray')
```

```python
F_RCombSin = abs(np.fft.fft2(RCombSin))
F_RCombSin1 = np.fft.fftshift(F_RCombSin)
plt.imshow(F_RCombSin1, cmap = 'gray')
```
