### Lunar landing scanned pictures: Line removal

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])
```

```python
moon = img.imread('moon.gif')
gray= rgb2gray(moon)
Moon = gray
plt.imshow(Moon, cmap = 'gray')
print(Moon.shape)
```

```python
Moon1 = abs(np.fft.fft2(Moon))
Moon2 = np.log(abs(np.fft.fftshift(Moon1)))
plt.imshow(Moon2, cmap ='gray')
print(Moon2.shape)
```

### Mask to be used to remove the vertical line from the original image

```python
mask = img.imread('maskmoon.jpg')
gray1 = rgb2gray(mask)
Mask = gray1
plt.imshow(Mask, cmap = 'gray')
print(Mask.shape)
```

```python
Mask1 =np.fft.fftshift(gray1)
print(Mask1.shape)
```

### Final image with the vertical line removed

```python
FI = np.fft.fft2(gray)
final = FI*Mask1
IRA = np.fft.fft2(final)
FImage = abs(IRA)
plt.imshow(FImage,cmap='gray')
```
