### Convolution theorem redux

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.image as mpimg
from scipy.signal import convolve2d
```

### Fourier transform of two dots

```python
def rgb2gray(rgb):
    return np.dot(rgb[...,:3], [0.299,0.587,0.114])

image = mpimg.imread('2dots.bmp')
gray= rgb2gray(image)
Twodots = gray
plt.imshow(Twodots, cmap = 'gray')
```

```python
F_Twodots = abs(np.fft.fft2(Twodots))
F_Twodots1 = np.fft.fftshift(F_Twodots)
plt.imshow(F_Twodots1, cmap = 'gray')
```

### Fourier transform of Gaussians of variance:

```python
nx=100
x=np.linspace(-1,1,nx)
y=np.linspace(-1,1,nx)
X,Y=np.meshgrid(x,y)
sigma = 0.2
G = np.exp(-((X + 0.5)**2+Y**2)/sigma**2) + np.exp(-((X - 0.5)**2+Y**2)/sigma**2)
plt.imshow(G,cmap='gray')
plt.show()

FI = np.fft.fft2(G)
FIS = abs(np.fft.fftshift(FI))
plt.imshow(FIS,cmap='gray')
plt.show()
```

### Random array of dirac deltas

```python
A = np.zeros((200,200))
i = 0
while i < 10:
    A[np.random.randint(0,200),np.random.randint(0,200)] = 1
    i+=1

plt.imshow(A, cmap = 'gray')
```

### Arbitrary 9x9 pattern

```python
d = np.zeros((9,9))
j = 0
k = 0 
while j < 9:
    while k <9:
        d[j,k] = np.random.randint(0,2)
        k +=1
    k=0
    j += 1

plt.imshow(d, cmap = 'gray')
```

```python
def conv2(x, y, mode = 'same'):
    return np.rot90(convolve2d(np.rot90(x, 2), np.rot90(y, 2), mode=mode), 2)
```

### Convolution of a random array of Dirac Deltas with an arbitrary 9x9 pattern

```python
F_Ad = abs(conv2(A,d))
plt.imshow(F_Ad, cmap ='gray')
```

### Array of equally spaced dots

```python
n=200
A1=np.zeros((n,n))
number=np.arange(100,200,20)
for n in number:
    A1[100][n]=1
    A1[100][-n]=1
    A1[n][100]=1
    A1[-n][100]=1
    
plt.imshow(A1, cmap='gray')
```

```python
F_A1 = abs(np.fft.fft2(A1))
F_A11 = np.fft.fftshift(F_A1)
plt.imshow(F_A11, cmap = 'gray')
```
